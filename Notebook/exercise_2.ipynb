## Calcular CE

```python
# CE es la suma de las lecturas de todos los periodos por día y frontera para Lecturas_parte1.csv
ce_diario = df_1.groupby(['Fecha', 'CodFronteraDDV'])['Lectura'].sum().reset_index()
ce_diario.rename(columns={'Lectura': 'CE'}, inplace=True)
```

```python
# Unimos df_full con df_mapeo usando CodFronteraDDV (y Fecha si aplica)
df_post = pd.merge(df_full, df_mapeo[['Fecha', 'CodFronteraDDV', 'ValorLBC']], 
                   on=['Fecha', 'CodFronteraDDV'], how='left')

# Ahora df_post tiene la columna ValorLBC correspondiente a cada CodFronteraDDV
df_post.head()
```

```python
# Unir con Mapeo para obtener LBC (Línea Base de Consumo)
datos_completos = pd.merge(ce_diario, df_mapeo, on=['Fecha', 'CodFronteraDDV'], how='left')
datos_completos.head()
```

```python
# Calculamos la desconexión verificada
# Desconexión = MAX(0, LBC - CE)
datos_completos['Desconexion_Verificada'] = datos_completos.apply(
    lambda row: max(0, row['ValorLBC'] - row['CE']), axis=1
)

datos_completos.head(10)
```

```python
# Agrupamos por Agente para obtener totales mensuales
# Extraer mes de la fecha
datos_completos['Fecha_dt'] = pd.to_datetime(datos_completos['Fecha'])
datos_completos['Anio_Mes'] = datos_completos['Fecha_dt'].dt.to_period('M')

# Agrupar por mes y agente
resultado_final = datos_completos.groupby(['Anio_Mes', 'AGENTE']).agg(
    Desconexion_Verificada_kWh=('Desconexion_Verificada', 'sum'),
    Total_Fronteras=('CodFronteraDDV', 'nunique')
).reset_index()
```

```python
resultado_final['Fecha'] = resultado_final['Anio_Mes'].astype(str)
resultado_final = resultado_final[['Fecha', 'AGENTE', 'Desconexion_Verificada_kWh', 'Total_Fronteras']]

# Ordenar por código de agente
resultado_final = resultado_final.sort_values('AGENTE')
```
